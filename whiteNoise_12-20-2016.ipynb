```julia
using Plots

function generate_data(n)
    epsilon_values = Array(Float64, n)
    for i = 1:n
        epsilon_values[i] = randn()
    end
    return epsilon_values
end

ts_length = 100
data = generate_data(ts_length)
plot(data, color="blue")

            
```

```julia
#exercise 1
function factorial4(n)
    agg = 1
    fact = 1    
    for i in 1:n
        fact = fact * agg
        agg = agg + 1
    end
    return fact
end    

factorial4(5)
```
