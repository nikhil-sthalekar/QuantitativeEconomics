```julia
using Distributions
```

```julia
# Defining AR1 int the same way as in the lecture
struct AR1_best{T <: Real}
    a::T
    b::T
    sigma::T
    phi::Distribution
end
```

```julia
function simulate(m::AR1_best, n::Integer, x0::Real)
    # initializing array of length n
    X = Array{Float64}(n)
    
    # defining first element as x0
    X[1] = x0
    
    # contructing the array
    for i in 1:(n - 1)
        # setting each value using the law of motion
        # X_t+1 = a*X_t + b + sigma*W_t+1 
        X[i + 1] = m.a * X[i] + m.b + m.sigma *rand(m.phi)
    end
    return X
end 
```

```julia
m = AR1_best(0.9, 1.0, 1.0, Beta(5, 5))
X = simulate(m, 1000, 0.0)
```
