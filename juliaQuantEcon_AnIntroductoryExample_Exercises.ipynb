```julia
#exercise 1
function factorial2(n)
    agg = 1
    fact = 1    
    for i in 1:n
        fact = fact * agg
        agg = agg + 1
    end
    return fact
end    

factorial2(5)
factorial(5)
```

```julia
#exercise 2
using Plots
function binomial_rx(n, p)
    #the binomial distribution with parameters n and p is the 
    #discrete probability distribution of the number of successes 
    #in a sequence of n independent yes/no experiments, 
    #each of which yields success with probability p.
    agg = 0
    Dist = rand(n)
    #iterates through number of trials, n
    for i in 1:n
        if Dist[i] < p
            agg += 1    # Or count += 1
        end
    end
    return agg
end

#doing some trials and looking at them in a histogram
trials = 10000
b = Array(Int, trials)
for j in 1:trials
    b[j] = binomial_rx(100, 0.5)
end

histogram(b)
```

```julia
#exercise 3: Monte Carlo Approximation of Pi
using Plots
n = 10000
U1 = Array{Float64}(n)
U2 = Array{Float64}(n)
for i in 1:n
    U1[i] = rand()
    U2[i] = rand()
end



agg = 0
B1 = Array{Float64}(n)
B2 = Array{Float64}(n)
for j in 1:n
    if (((U1[j] - 0.5)^2) + ((U2[j] - 0.5)^2)) < (0.25)
        agg += 1
        B1[j] = U1[j]
        B2[j] = U2[j]
    else 
        agg += 0
        B1[j] = 0.5
        B2[j] = 0.5
    end
end

p = 4 * (agg/n)

println(p)
```

```julia
#exercise 3: Approximation of Pi using Monte Carlo Method, Visualization

plot(
plot(U1,U2, seriestype = [:scatter]),
plot(B1,B2, seriestype = [:scatter])
)
```

```julia
#exercise 4: Gambling Device
flips = 10
a = Array{Int}(flips)

#0 is tails, 1 is heads
for i in 1:flips
    a[i] = rand(0:1)
end

for j in 3:flips
    agg = 0
    if a[j] == 1 && a[j-1] == 1 && a[j-2] == 1
        agg += 1
    else
        agg += 0
    end
end

if agg > 0
    println("pay up, one dollar!")
else
    println("woo, I didn't lose money!")
end

a

```

```julia
#exercise 5 and 6
using Plots
T = 200
alpha = 0.9
alpha1 = 0
alpha2 = 0.8
alpha3 = 0.98

function corrTimeSeries(Timesteps, a)
    X = Array{Float64}(Timesteps)
    epsilon = Array{Float64}(Timesteps)
    t = 1
    X[1] = 0
    for t in 1:Timesteps
        epsilon[t] = randn()
    end
   
    
    for t in 2: (Timesteps-1)
        X[t + 1] = a * X[t] + epsilon[t + 1]
        t += 1
    end
    return X
end





plot([corrTimeSeries(T, alpha),corrTimeSeries(T, alpha1),corrTimeSeries(T, alpha2),corrTimeSeries(T, alpha3)]
)

```
