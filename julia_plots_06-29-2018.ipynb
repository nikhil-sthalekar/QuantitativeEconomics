```julia
import Plots
```

```julia
# 3D plot
Plots.plotlyjs()

n = 50
x = linspace(-3, 3, n)
y = x

z = Array{Float64}(n, n)
ff(x, y) = cos(x^2 + y^2 + 2*x*y) / (1 + x^2 + y^2)
for i in 1:n
    for j in 1:n
        z[j, i] = ff(x[i], y[j])
    end
end

Plots.surface(x, y, z')
```

```julia
function d(x, n)
    val = 0
    for k in 1:(n * 2^n)
        if (k - 1) / 2^n <= x < k / 2^n
            val += (k - 1) / 2^n
        end
    end
    if x >= n
        val += n
    end
    return val
end
```

```julia
N = [1, 2, 3, 4, 5]
X = linspace(0, 10, 50)
vals = []
for n in N
    push!(vals , [d(x, n) for x in X])
end

push!(vals, X)
labels
Plots.plot(X,
           vals)
```
