```julia
using Plots
ts_length = 100
e_values = randn(ts_length)
plot(e_values, color="blue")
```

```julia
# For Loops
ts_length = 100
e_value = Array{Float64}(ts_length)
for i in 1:ts_length
    e_values[i] = randn()
end
plot(e_values, color = "blue")
```

```julia
# While Loops
t_length = 100
t_values = Array{Float64}(t_length)
i = 1
while i <= t_length
    t_values[i] = randn()
    i = i + 1
end
plot(t_values, color = "blue")
```

```julia
ts_length = 100
ϵ_values = Array{Float64}(ts_length)
i = 1
while true
    ϵ_values[i] = randn()
    i = i + 1
    if i > ts_length
        break
    end
end
plot(ϵ_values, color="blue")
```

```julia
# User Defined Functions
function generate_data(n)
    ϵ_values = Array{Float64}(n)
    for i = 1:n
        ϵ_values[i] = randn()
    end
    return ϵ_values
end

ts_length = 100
data = generate_data(ts_length)
plot(data, color="blue")
```

```julia
# An Example of a more useful Function
using Distributions

function plot_histogram(distribution, n)
    ϵ_values = rand(distribution, n)  # n draws from distribution
    histogram(ϵ_values)
end

lp = Laplace()
plot_histogram(lp, 500)
```
